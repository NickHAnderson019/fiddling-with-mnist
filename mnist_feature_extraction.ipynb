# Feature Extraction

### Let's look under the hood and see what exactly each filter in out CNN is outputing. 

First, we need to load our model and our test data.

```python
from keras.datasets import mnist
from keras.utils.np_utils import to_categorical
import numpy as np
import matplotlib.pyplot as plt
import cv2
import math
```

```python
# LOAD DATA
(_, _), (x_test, y_test) = mnist.load_data() # ignore x_train, y_train
```

```python
# PRE-PROCESSING
y_cat_test = to_categorical(y_test,10)
x_test = x_test/255
x_test = x_test.reshape(10000,28,28,1)
```

```python
from keras.models import load_model
model = load_model("models/MNIST_CNN_Model.h5")
model.summary()
```

### CNN Layer outputs

Now that we hace our model, we can get the outputs at each layer for a test image. 

```python
from keras import backend as K

input_shape = x_test.shape

inp = model.input                                           # input placeholder
outputs = [layer.output for layer in model.layers]          # all layer outputs
functor = K.function([inp, K.learning_phase()], outputs )   # evaluation function
```

```python
# Testing - this is our test image
test = x_test[0][np.newaxis,...]
test_img = test.reshape(28,28)
plt.imshow(test_img,'gray')
plt.title("Test Image - 7")
plt.show()
```

```python
# Get output from each layer
layer_outs = functor([test, 1.])

# Let's check that the output at our last layer (the label) has argmax of 7.
layer_outs[-1][0].argmax()
```

```python
# Get output of first conv layer
first_layer_out = layer_outs[0][0]
first_layer_out.shape
```

```python
# Get output of first conv layer first filter
first_layer_filter_one_out = first_layer_out[:,:,0]
first_layer_filter_one_out.shape
```

```python
# Plot output of layer filter
first_layer_filter_one_out_img = first_layer_filter_one_out.reshape(first_layer_filter_one_out.shape[0],first_layer_filter_one_out.shape[1])
plt.imshow(first_layer_filter_one_out_img,'gray')
plt.title("First Layer, First Filter")
plt.show()
```

Now that we know how to get the output of a single filter in a single layer in the CNN, we can expand that to include all convolution, pooling and dropout layers in the CNN.

## Let's write a function

We can now write a function to get the output for all filters in a given layer for multiple test images.

```python
def get_features_visual(layer, num_images):
    num_test_images = x_test.shape[0]
    test_index_array = np.linspace(0,num_test_images-1,num_test_images, dtype=np.uint32)
    test_indices = np.random.choice(test_index_array,num_images)
    
    for ind in test_indices:
        # Test image
        test = x_test[ind][np.newaxis,...]
        test_img = test.reshape(28,28)

        # Get output from each layer
        layer_outs = functor([test, 1.])

        # number of filter for layer
        num_filters = layer_outs[layer].shape[3]
        num_cols = 16

        fig = plt.figure(figsize=(15,2))
        for i in range(num_filters):
            filter_out_img = layer_outs[layer][0,:,:,i]
            plt.subplot(math.ceil(num_filters/num_cols), num_cols, i+1)
            plt.xticks([])
            plt.yticks([])
            plt.grid(False)
            plt.imshow(filter_out_img, 'gray')
        fig.subplots_adjust(wspace=0, hspace=0)
```

### Visualise
We can now visualise the features extracted in layers 0-4 below for num_images number of random test_images.

```python
np.random.seed(1) # change this to get new random images

# Get plots of num_images amount of random images
num_images = 5
layer = 0 # conv, max pooling and dropout are layers 0-4
get_features_visual(layer, num_images)
```

## Some quick notes

Looking at the first filter for the conv layers, we see that it is almost like a light is shining from below on the image. For the second filter, it seems that a light is shining form above. Both of these have the function of extracting the horizontal lines in the image.

As you would expect, the image becomes more and more distorted as you move deeper into the CNN due to the convolutional, pooling and dropout layers which all reduce dimensionality of the input.
